```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv(
    "/Users/hiteshjindal/Downloads/Projects/Recommendation system/Dataset/archive/tmdb_5000_credits.csv")
movies_df = pd.read_csv(
    "/Users/hiteshjindal/Downloads/Projects/Recommendation system/Dataset/archive/tmdb_5000_movies.csv")
```

```python
credits.head()
movies_df.head()
```

```python
print("Credits:",credits.shape)
print("Movies Dataframe:",movies_df.shape)
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})
```

```python
movies_df_merge = movies_df.merge(credits_column_renamed, on="id")
```

```python
movies_df_merge.head()
```

```python
movies_cleaned_df = movies_df_merge.drop(columns = ["homepage","title_x","title_y","status",
                                                    "production_companies"])
```

```python
movies_cleaned_df.head()
movies_cleaned_df.info()
```

# use weighted average for each movie's average rating
                    W = weighted rating = (Rv + Cm)/(v*m)
                    v= no. of votes for the movie
                    R = average for the movie as a number from 0 to 10
                    m = minimum votes required to be in top 250 list
                    C = the mean vote across the whole report

```python
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(.70)
```

```python
movies_cleaned_df['weighted_average'] = (R * v + C * m)/(v + m)
movies_cleaned_df.head()
```

```python
movie_sorted_ranking= movies_cleaned_df.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1= sns.barplot(x=weight_average['weighted_average'].head(10),y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title("Best Movies by average votes",weight='bold')
plt.xlabel("Weighted Average Score",weight='bold')
plt.ylabel("Movie Title",weight='bold')
plt.savefig("best_movie.png")
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
axis2= sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10))

plt.title("Best Movies by popularity",weight='bold')
plt.xlabel("Popularity",weight='bold')
plt.ylabel("Movie Title",weight='bold')
plt.savefig("best_movie.png")
```

# Recommendation based on scaled weighted average and popularity score(priority is given 50% to both)



```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']]) ## scaled 
# value 0 to 1

movie_normalized_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_average','normalized_popularity']] = movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_average']*0.5 + movies_cleaned_df[
    'normalized_popularity']*0.5
movies_scored_df = movies_cleaned_df.sort_values('score',ascending=False)
movies_scored_df[['original_title','normalized_average','normalized_popularity','score']].head(20)
```

```python
scored_df=movies_cleaned_df.sort_values('score',ascending=False)
plt.figure(figsize=(16,6))
ax= sns.barplot(x=scored_df['score'].head(10),y=scored_df['original_title'].head(10))

plt.title("Best Rated & Most popular blend",weight='bold')
plt.xlabel("scores",weight='bold')
plt.ylabel("Movie Title",weight='bold')
plt.savefig("best_movie.png")
```
